```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url1 = 'https://raw.githubusercontent.com/Jessie-Grewal/Neur265/main/alzheimers_pathology_final_project.csv'
url2 = 'https://raw.githubusercontent.com/Jessie-Grewal/Neur265/main/alzheimers_IDs_final_project.csv'

Table1 = pd.read_csv(url1)
Table2 = pd.read_csv(url2)
```

```python
merged_table = pd.merge(Table1, Table2[['Donor ID', 'Cognitive Status', 'Sex','CERAD score', 'Last CASI Score', 'APOE4 Status', 'Last MMSE Score']], on = 'Donor ID')
```

```python
merged_table
```

```python
merged_table.to_csv('merged_table.csv', index=False)
```

```python
merged_table['total Iba1 positive area_Layer1'].hist(bins=10)
plt.xlabel('Total IBA1 Positive Area_Layer1')
plt.ylabel('Frequency')
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=5, figsize=(10, 5))

# Plot histograms of the 5 columns in the corresponding subplots

merged_table['total Iba1 positive area_Layer1'].hist(ax=ax[0],bins=10)
ax[0].set_title('Histogram of Layer 1')

merged_table['total Iba1 positive area_Layer2'].hist(ax=ax[1], bins=10)
ax[1].set_title('Histogram of Layer 2')

merged_table['total Iba1 positive area_Layer3'].hist(ax=ax[2], bins=10)
ax[2].set_title('Histogram of Layer 3')

merged_table['total Iba1 positive area_Layer4'].hist(ax=ax[3], bins=10)
ax[3].set_title('Histogram of layer 4')

merged_table['total Iba1 positive area_Layer5-6'].hist(ax=ax[4], bins=10)
ax[4].set_title('Histogram of Layer 5-6')

# Add overall title and adjust spacing
fig.suptitle('Histograms of 5 variables', fontsize=16)
fig.tight_layout()

# Show the plot
plt.show()
```
