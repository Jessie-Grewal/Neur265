```python
# import numpy and matplotlib modules 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import acc_hits, acc_fa, lc_hits, and lc_fa as numpy arrays

url1 = 'https://raw.githubusercontent.com/Jessie-Grewal/Neur265/main/acc_fa_hw6%20(1).csv'
url2 = 'https://raw.githubusercontent.com/Jessie-Grewal/Neur265/main/acc_hits_hw6%20(1).csv'
url3 = 'https://raw.githubusercontent.com/Jessie-Grewal/Neur265/main/lc_fa_hw6%20(1).csv'
url4 = 'https://raw.githubusercontent.com/Jessie-Grewal/Neur265/main/lc_hits_hw6%20(1).csv'

acc_fa = np.loadtxt(url1, delimiter = ',')
acc_hits = np.loadtxt(url2, delimiter = ',')
lc_fa = np.loadtxt(url3, delimiter = ',')
lc_hits = np.loadtxt(url4, delimiter = ',')
```

```python
# print the shape of each array

print(acc_fa.shape)
print(acc_hits.shape)
print(lc_fa.shape)
print(lc_hits.shape)
```

```python
# create time variable that has the same length of the four arrays with equally spaced values between zero and four

time = np.linspace(0,4,acc_fa.shape[1])
len(time)
```

```python
# create an expression that determines the sampling rate of the data given that there are 8001 samples and each sample is 4 seconds 

sample_rate = acc_fa.shape[1]/4
print(sample_rate)
```

```python
# plot the first trial of acc_hits in red, and the first trial of lc_hits in the same plot in blue in one subplot. In a second subplot, plot the first trial of acc_fa in red, and the first trial of lc_fa in blue.

fig, ax = plt.subplots(1,2, figsize = (15,5))
ax[0].plot(time, acc_hits[0], color = 'r')
ax[0].plot(time, lc_hits[0], color = 'b')
ax[0].set_xlabel('Time [s]')              # Label the x-axis as time.
ax[0].set_ylabel('Voltage [\mu V]')       # Label the y-axis as voltage.
ax[1].plot(time, acc_fa[0], color = 'r')
ax[1].plot(time, lc_fa[0], color = 'b')
ax[1].set_xlabel('Time [s]')              # Label the x-axis as time.
ax[1].set_ylabel('Voltage [\mu V]')       # Label the y-axis as voltage.
ax[0].set_title('Hits Voltage over Time')
ax[1].set_title('False Alarm Voltage over Time')
```

By looking at the number of peaks in the signal and dividing by the total length of the signal (4 seconds), the dominant oscillation in the ACC LFP appears to be about 10 peaks/4 seconds, or 2.5 Hz for Hits. The dominant oscillation in the LC LFP appears to be 14 peaks/4 seconds, or 3.5 Hz for Hits. It appears that in some cases as in the time period between 2.5-2.8 seconds in Hits Voltage over Time, the peaks of the LC correspond to the troughs of the ACC. This is similar in the time period between approximately 1.7-2.0 seconds in the False Alarm Voltage over time plot as the peaks of the LC correspond to trough of ACC. Another relationship between the two brain areas in Hits and False alarms over time is that bursts of activity in the LC generally precede a large voltage change in the ACC throughout the trial. 


```python
# Compute the coherence for acc_hits and lc_hits

# Compute the Fourier transforms
x = acc_hits[1,:] - np.mean(acc_hits[1,:])  # Define one time series,
y = lc_hits[1,:] - np.mean(lc_hits[1,:])  # ... and another.

xf = np.array([np.fft.rfft(x-x.mean()) for x in acc_hits])
yf = np.array([np.fft.rfft(y-y.mean()) for y in lc_hits])

# Compute the spectra
T = time[-1]
dt = time[1] - time[0]              # Define the sampling interval.
scale = 2 * dt**2 / T
Sxx = scale * (xf * xf.conj()).mean(0)  # Spectrum of E1 trials
Syy = scale * (yf * yf.conj()).mean(0)  # ... and E2 trials
Sxy = scale * (xf * yf.conj()).mean(0)  # ... and the cross spectrum

# Compute the Coherence
cohr = np.abs(Sxy)/(np.sqrt(Sxx)* np.sqrt(Syy))
N = acc_hits.shape[1]

f = np.fft.rfftfreq(N, dt)
plt.plot(f, cohr.real)
plt.xlim([0,20])
plt.ylim([0,1])
plt.xlabel('Frequency [Hz]')
plt.ylabel('Coherence')
plt.title('Coherence between ACC_Hits and LC_Hits')
```

The coherence is close to 1, indicating a strong coherence around the 1.5 Hz band and drops to approximately 0.4 between the 2.5-4 Hz bands. The coherence increases and reaches a value close to 1.0 at about 5 Hz and steadily decreases as the frequency increases. 

```python
# calculate coherence for false alarms 

# Compute the Fourier transforms
x_fa = acc_fa[1,:] - np.mean(acc_fa[1,:])  # Define one time series,
y_fa = lc_fa[1,:] - np.mean(lc_fa[1,:])  # ... and another.
xf_fa = np.array([np.fft.rfft(x_fa-x_fa.mean()) for x_fa in acc_fa])
yf_fa = np.array([np.fft.rfft(y_fa-y_fa.mean()) for y_fa in lc_fa])

# Compute the spectra
T = time[-1]
dt = time[1] - time[0]              # Define the sampling interval.
scale = 2 * dt**2 / T
Sxx = scale * (xf_fa * xf_fa.conj()).mean(0)  # Spectrum of E1 trials
Syy = scale * (yf_fa * yf_fa.conj()).mean(0)  # ... and E2 trials
Sxy = scale * (xf_fa * yf_fa.conj()).mean(0)  # ... and the cross spectrum

# Compute the Coherence
cohr2 = np.abs(Sxy)/(np.sqrt(Sxx)* np.sqrt(Syy))
N2 = acc_fa.shape[1]

# plot the results from this step (coherence in false alarms) across frequency

f_fa = np.fft.rfftfreq(N2, dt)
plt.plot(f_fa, cohr2.real)
plt.xlim([0,20])
plt.ylim([0,1])
plt.xlabel('Frequency [Hz]')
plt.ylabel('Coherence')
plt.title('Coherence between ACC and LC')
```

```python
# plot the coherence in hits and in false alarms across frequency in the same plot 
plt.plot(f, cohr.real, color = 'r')
plt.plot(f_fa, cohr2.real, color = 'b')
plt.xlim([0,20])
plt.ylim([0,1])
plt.xlabel('Frequency [Hz]')
plt.ylabel('Coherence')
plt.title('Coherence between ACC and LC')
```

The coherence between the ACC and LC follows a similar trend across frequencies for hits and false alarms. The coherence appears to be greatest and closest to 1.0 at about 1.5 Hz before dropping significantly at about 2.5 Hz before rising close to 1 at about 5 Hz. From there, the coherence appears to steadily decrease as the frequency increases for hits and false alarms. The coherence is generally higher during hits than in false alarms. Given that higher coherence means more communication between ACC and LC, these two brain areas are communicating more during hits since the coherence values at most frequencies are higher for hits relative to false alarms.
