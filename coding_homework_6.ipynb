```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
url1 = 'https://raw.githubusercontent.com/Jessie-Grewal/Neur265/main/acc_fa_hw6.csv'
url2 = 'https://raw.githubusercontent.com/Jessie-Grewal/Neur265/main/acc_hits_hw6.csv'
url3 = 'https://raw.githubusercontent.com/Jessie-Grewal/Neur265/main/lc_fa_hw6.csv'
url4 = 'https://raw.githubusercontent.com/Jessie-Grewal/Neur265/main/lc_hits_hw6.csv'

acc_fa = np.loadtxt(url1, delimiter = ',')
acc_hits = np.loadtxt(url2, delimiter = ',')
lc_fa = np.loadtxt(url3, delimiter = ',')
lc_hits = np.loadtxt(url4, delimiter = ',')
```

```python
print(acc_fa.shape)
print(acc_hits.shape)
print(lc_fa.shape)
print(lc_hits.shape)
```

```python
time = np.linspace(0,4,acc_fa.shape[1])
len(time)
```

```python
sample_rate = acc_fa.shape[1]/4
print(sample_rate)
```

```python
fig, ax = plt.subplots(1,2, figsize = (15,5))
ax[0].plot(time, acc_hits[0], color = 'r')
ax[0].plot(time, lc_hits[0], color = 'b')
ax[0].set_xlabel('Time [s]')              # Label the x-axis as time.
ax[0].set_ylabel('Voltage [\mu V]')       # Label the y-axis as voltage.
ax[1].plot(time, acc_fa[0], color = 'r')
ax[1].plot(time, lc_fa[0], color = 'b')
ax[1].set_xlabel('Time [s]')              # Label the x-axis as time.
ax[1].set_ylabel('Voltage [\mu V]')       # Label the y-axis as voltage.
ax[0].set_title('Hits Voltage over Time')
ax[1].set_title('False Alarm Voltage over Time')
```

```python
# Compute the Fourier transforms
x = acc_hits[1,:] - np.mean(acc_hits[1,:])  # Define one time series,
y = lc_hits[1,:] - np.mean(lc_hits[1,:])  # ... and another.

xf = np.array([np.fft.rfft(x-x.mean()) for x in acc_hits])
yf = np.array([np.fft.rfft(y-y.mean()) for y in lc_hits])

# Compute the spectra
T = time[-1]
dt = time[1] - time[0]              # Define the sampling interval.
scale = 2 * dt**2 / T
Sxx = scale * (xf * xf.conj()).mean(0)  # Spectrum of E1 trials
Syy = scale * (yf * yf.conj()).mean(0)  # ... and E2 trials
Sxy = scale * (xf * yf.conj()).mean(0)  # ... and the cross spectrum

# Compute the Coherence
cohr = np.abs(Sxy)/(np.sqrt(Sxx)* np.sqrt(Syy))
N = acc_hits.shape[1]

f = np.fft.rfftfreq(N, dt)
plt.plot(f, cohr.real)
plt.xlim([0,20])
plt.ylim([0,1])
plt.xlabel('Frequency [Hz]')
plt.ylabel('Coherence')
plt.title('Coherence between ACC and LC')
```

```python
# Compute the Fourier transforms
x2 = acc_fa[1,:] - np.mean(acc_fa[1,:])  # Define one time series,
y2 = lc_fa[1,:] - np.mean(lc_fa[1,:])  # ... and another.
xf_fa = np.array([np.fft.rfft(x2-x2.mean()) for x2 in acc_fa])
yf_fa = np.array([np.fft.rfft(y2-y2.mean()) for y2 in lc_fa])

# Compute the spectra
T = time[-1]
dt = time[1] - time[0]              # Define the sampling interval.
scale = 2 * dt**2 / T
Sxx = scale * (xf_fa * xf_fa.conj()).mean(0)  # Spectrum of E1 trials
Syy = scale * (yf_fa * yf_fa.conj()).mean(0)  # ... and E2 trials
Sxy = scale * (xf_fa * yf_fa.conj()).mean(0)  # ... and the cross spectrum

# Compute the Coherence
cohr2 = np.abs(Sxy)/(np.sqrt(Sxx)* np.sqrt(Syy))
N2 = acc_fa.shape[1]

f2 = np.fft.rfftfreq(N2, dt)
plt.plot(f2, cohr2.real)
plt.xlim([0,20])
plt.ylim([0,1])
plt.xlabel('Frequency [Hz]')
plt.ylabel('Coherence')
plt.title('Coherence between ACC and LC')
```
